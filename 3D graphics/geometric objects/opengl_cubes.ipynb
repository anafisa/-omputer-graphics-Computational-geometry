### Task 1: multi-coloured cube

```python
from OpenGL.GL import *
from OpenGL.GLU import *
from OpenGL.GLUT import *

def init():
    glEnable(GL_DEPTH_TEST)
    glClearColor(0.5, 0.5, 0.5, 1.0) 
    gluOrtho2D(-1.0, 1.0, -1.0, 1.0) 


def keyboardkeys(key, x, y):
    if key == b'\x1b':
        sys.exit(0)
    glutPostRedisplay()         

    
def draw(*args, **kwargs):
    glClear(GL_COLOR_BUFFER_BIT | GL_DEPTH_BUFFER_BIT) 
    glRotated(0.125,1,1,1)

    glBegin(GL_QUADS)

    glColor3f(1, 0, 0)
    glVertex3f(0.5, -0.5, 0.5 )
    glVertex3f( 0.5, 0.5, 0.5 )
    glVertex3f(-0.5, 0.5, 0.5 )
    glVertex3f(-0.5, -0.5, 0.5)
    glEnd()
   

    glBegin(GL_QUADS)
    glColor3f(0, 1, 0)
    glVertex3f( 0.5, -0.5, -0.5 )
    glVertex3f( 0.5,  0.5, -0.5 )
    glVertex3f( 0.5,  0.5,  0.5 )
    glVertex3f( 0.5, -0.5,  0.5 )
    glEnd()
    
    
    glBegin(GL_QUADS)
    glColor3f(0, 0, 1)
    glVertex3f( -0.5, -0.5,  0.5 )
    glVertex3f( -0.5,  0.5,  0.5 )
    glVertex3f( -0.5,  0.5, -0.5 )
    glVertex3f( -0.5, -0.5, -0.5 )
    glEnd()
    
    
    glBegin(GL_QUADS)
    glColor3f( 0, 1, 1)
    glVertex3f(  0.5,  0.5,  0.5 )
    glVertex3f(  0.5,  0.5, -0.5 )
    glVertex3f( -0.5,  0.5, -0.5 )
    glVertex3f( -0.5,  0.5,  0.5 )
    glEnd()

    
    glBegin(GL_QUADS)
    glColor3f(1, 0, 1)
    glVertex3f(  0.5, -0.5, -0.5 )
    glVertex3f(  0.5, -0.5,  0.5 )
    glVertex3f( -0.5, -0.5,  0.5 )
    glVertex3f( -0.5, -0.5, -0.5 )
    glEnd()
    
    
    glBegin(GL_QUADS)
    glColor3f(1, 1, 0)
    glVertex3f(  0.5, -0.5, -0.5 )
    glVertex3f(  0.5,  0.5, -0.5 )
    glVertex3f( -0.5,  0.5, -0.5 )
    glVertex3f( -0.5, -0.5, -0.5 )
    glEnd()

    
    glutSwapBuffers()        
    glutPostRedisplay()         

glutInitDisplayMode(GLUT_DOUBLE | GLUT_RGB | GLUT_DEPTH)
glutInitWindowSize(800, 600)
glutInitWindowPosition(50, 50)
glutInit(sys.argv)
glutCreateWindow(b"OpenGL Cube!")
glutDisplayFunc(draw)

glutKeyboardFunc(keyboardkeys)
init()
glutMainLoop()
```

## Task 2: small cubes

```python
from OpenGL.GL import *
from OpenGL.GLU import *
from OpenGL.GLUT import *

def init():
    glEnable(GL_DEPTH_TEST)
    glClearColor(0.5, 0.5, 0.5, 1.0) 
    gluOrtho2D(-1.0, 1.0, -1.0, 1.0) 


def keyboardkeys(key, x, y):
    if key == b'\x1b':
        sys.exit(0)
    glutPostRedisplay()         

    
def cube(x, y, z, size):   
    
    x -= size/2
    y -= size/2
    z -= size/2
    
    glColor3f(1, 0, 0)
    glVertex3f(x, y, z)
    glVertex3f(x + size, y, z)
    glVertex3f(x + size, y, z + size)
    glVertex3f(x, y, z + size)


    glColor3f(0, 1, 0)
    glVertex3f(x, y - size, z )
    glVertex3f(x + size, y - size, z)
    glVertex3f(x + size, y - size, z + size)
    glVertex3f(x, y - size , z + size)
    
    
    glColor3f(0, 0, 1)
    glVertex3f(x, y, z)
    glVertex3f(x, y, z + size)
    glVertex3f(x, y - size, z + size)
    glVertex3f(x, y - size,z)
    
    
    glColor3f( 0, 1, 1)
    glVertex3f(x + size, y, z)
    glVertex3f(x + size, y, z + size)
    glVertex3f(x + size, y - size, z + size)
    glVertex3f(x + size, y - size,z)


    glColor3f(1, 0, 1)
    glVertex3f(x + size, y, z)
    glVertex3f(x + size , y-size, z)
    glVertex3f(x, y - size, z)
    glVertex3f(x, y, z)
    
    
    glColor3f(1, 1, 0)
    glVertex3f(x+size, y, z + size)
    glVertex3f(x+size, y-size, z + size)
    glVertex3f(x, y - size, z + size)
    glVertex3f(x, y, z + size)


    
def draw(*args, **kwargs):
    glClear(GL_COLOR_BUFFER_BIT | GL_DEPTH_BUFFER_BIT) 
    glRotated(0.125,1,1,1)
   
    glBegin(GL_QUADS)
    cube(0.4, 0.4, 0.4, 0.2)
    glEnd()
    
    glBegin(GL_QUADS)
    cube(-0.4, 0.4, 0.4, 0.2)
    glEnd()
    
    glBegin(GL_QUADS)
    cube(0.4, -0.4, 0.4, 0.2)
    glEnd()
    
    glBegin(GL_QUADS)
    cube(0.4, 0.4, -0.4, 0.2)
    glEnd()
    
    glBegin(GL_QUADS)
    cube(-0.4, -0.4, 0.4, 0.2)
    glEnd()
    
    glBegin(GL_QUADS)
    cube(-0.4, 0.4, -0.4, 0.2)
    glEnd()
    
    glBegin(GL_QUADS)
    cube(0.4, -0.4, -0.4, 0.2)
    glEnd()
    
    glBegin(GL_QUADS)
    cube(-0.4, -0.4, -0.4, 0.2)
    glEnd()
   
    glutSwapBuffers()        
    glutPostRedisplay()         

glutInitDisplayMode(GLUT_DOUBLE | GLUT_RGB | GLUT_DEPTH)
glutInitWindowSize(800, 600)
glutInitWindowPosition(50, 50)
glutInit(sys.argv)
glutCreateWindow(b"OpenGL Cube!")
glutDisplayFunc(draw)

glutKeyboardFunc(keyboardkeys)
init()
glutMainLoop()
```

## Task 4: cross od cubes

```python
from OpenGL.GL import *
from OpenGL.GLU import *
from OpenGL.GLUT import *

def init():
    glEnable(GL_DEPTH_TEST)
    glClearColor(0.5, 0.5, 0.5, 1.0) 
    gluOrtho2D(-1.0, 1.0, -1.0, 1.0) 


def keyboardkeys(key, x, y):
    if key == b'\x1b':
        sys.exit(0)
    glutPostRedisplay()         

    
def cube(x, y, z, size):   
    
    x -= size/2
    y -= size/2
    z -= size/2
    
    glVertex3f(x, y, z)
    glVertex3f(x + size, y, z)
    glVertex3f(x + size, y, z + size)
    glVertex3f(x, y, z + size)

    glVertex3f(x, y - size, z )
    glVertex3f(x + size, y - size, z)
    glVertex3f(x + size, y - size, z + size)
    glVertex3f(x, y - size , z + size)
    
    
    glVertex3f(x, y, z)
    glVertex3f(x, y, z + size)
    glVertex3f(x, y - size, z + size)
    glVertex3f(x, y - size,z)
    
    
    glVertex3f(x + size, y, z)
    glVertex3f(x + size, y, z + size)
    glVertex3f(x + size, y - size, z + size)
    glVertex3f(x + size, y - size,z)


    glVertex3f(x + size, y, z)
    glVertex3f(x + size , y-size, z)
    glVertex3f(x, y - size, z)
    glVertex3f(x, y, z)
    
    
    glVertex3f(x+size, y, z + size)
    glVertex3f(x+size, y-size, z + size)
    glVertex3f(x, y - size, z + size)
    glVertex3f(x, y, z + size)


    

def draw(*args, **kwargs):
    glClear(GL_COLOR_BUFFER_BIT | GL_DEPTH_BUFFER_BIT) 
    glRotated(0.125,1,1,1)
   
    k = 0.2
    for i in range(5):
        glBegin(GL_QUADS)
        glColor3f(1, 0, 0)
        cube(k, 0, 0, 0.2)
        glEnd()
        k += 0.2
        
    k = 0.2
    for i in range(5):
        glBegin(GL_QUADS)
        glColor3f(0, 1, 1)
        cube(-k, 0, 0, 0.2)
        glEnd()
        k += 0.2
    
    k = 0.2
    for i in range(5):
        glBegin(GL_QUADS)
        glColor3f(0, 1, 0)
        cube(0, k, 0, 0.2)
        glEnd()
        k += 0.2
        
    k = 0.2
    for i in range(5):
        glBegin(GL_QUADS)
        glColor3f(1, 0, 1)
        cube(0, -k, 0, 0.2)
        glEnd()
        k += 0.2
   
    k = 0.2
    for i in range(5):
        glBegin(GL_QUADS)
        glColor3f(0, 0, 1)
        cube(0, 0, k, 0.2)
        glEnd()
        k += 0.2
        
    k = 0.2
    for i in range(5):
        glBegin(GL_QUADS)
        glColor3f(1, 1, 0)
        cube(0, 0, -k, 0.2)
        glEnd()
        k += 0.2
    

    
    glutSwapBuffers()        
    glutPostRedisplay()         

glutInitDisplayMode(GLUT_DOUBLE | GLUT_RGB | GLUT_DEPTH)
glutInitWindowSize(800, 600)
glutInitWindowPosition(50, 50)
glutInit(sys.argv)
glutCreateWindow(b"OpenGL Cube!")
glutDisplayFunc(draw)

glutKeyboardFunc(keyboardkeys)
init()
glutMainLoop()
```
