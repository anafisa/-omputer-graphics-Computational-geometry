## Task 1: cone 

```python
from OpenGL.GL import *
from OpenGL.GLU import *
from OpenGL.GLUT import *
from math import cos, sin, pi

def init():
    glEnable(GL_DEPTH_TEST)
    glClearColor(0.5, 0.5, 0.5, 1.0) 
    gluOrtho2D(-1.0, 1.0, -1.0, 1.0) 

def keyboardkeys(key, x, y):
    if key == b'\x1b':
        sys.exit(0)
    glutPostRedisplay()         

def draw(*args, **kwargs):
    glClear(GL_COLOR_BUFFER_BIT | GL_DEPTH_BUFFER_BIT) 
    glRotated(0.125,1,1,1)
    R = 0.5
    
    glBegin(GL_TRIANGLE_FAN)
    glColor3f(1, 1, 1)
    glVertex3d(0, 0, 0)
    for i in range(21):
        glVertex3d(R*cos(i*2*pi/20), R*sin(i*2*pi/20), 0)
    glEnd()
    
    glBegin(GL_TRIANGLE_FAN)
    glColor3f(1, 1, 1)
    glVertex3d(0, 0, 1)
    for i in range(21):
        glVertex3d(R*cos(i*2*pi/20), R*sin(i*2*pi/20), 0)
    glEnd()

    glutSwapBuffers()           
    glutPostRedisplay()        

glutInitDisplayMode(GLUT_DOUBLE | GLUT_RGB | GLUT_DEPTH)
glutInitWindowSize(800, 600)
glutInitWindowPosition(50, 50)
glutInit(sys.argv)
glutCreateWindow(b"OpenGL")
glutDisplayFunc(draw)
glutKeyboardFunc(keyboardkeys)
init()
glutMainLoop()
```

## Task 6: sphere

```python
from OpenGL.GL import *
from OpenGL.GLU import *
from OpenGL.GLUT import *
from math import cos, sin, pi

def init():
    glEnable(GL_DEPTH_TEST)
    glClearColor(0.5, 0.5, 0.5, 1.0) 
    gluOrtho2D(-1.0, 1.0, -1.0, 1.0) 

def keyboardkeys(key, x, y):
    if key == b'\x1b':
        sys.exit(0)
    glutPostRedisplay()         

def draw(*args, **kwargs):
    glClear(GL_COLOR_BUFFER_BIT | GL_DEPTH_BUFFER_BIT) 
    glRotated(0.125,1,1,1)
    R = 0.5
        
    glBegin(GL_QUAD_STRIP)
    glColor3f(1, 1, 1)
    for j in range(-9, 8):
        z = R*sin(pi*j/18)
        r = R*cos(pi*j/18)
        for i in range(21):
            glVertex3d(R*cos(i*2*pi/20), R*sin(i*2*pi/20), 0)
            glVertex3d(r*cos((i+1)*2*pi/20), r*sin((i+1)*2*pi/20), z)
    glEnd()

    glutSwapBuffers()          
    glutPostRedisplay()        

glutInitDisplayMode(GLUT_DOUBLE | GLUT_RGB | GLUT_DEPTH)
glutInitWindowSize(800, 600)
glutInitWindowPosition(50, 50)
glutInit(sys.argv)
glutDisplayFunc(draw)
glutKeyboardFunc(keyboardkeys)
init()
glutMainLoop()
```

```python
from math import *
from OpenGL.GL import *
from OpenGL.GLU import *
from OpenGL.GLUT import *

def init():
    glEnable(GL_DEPTH_TEST)
    glClearColor(0.5, 0.5, 0.5, 1.0) 
    gluOrtho2D(-1.0, 1.0, -1.0, 1.0)

def keyboardkeys(key, x, y):
    if key == b'\x1b':
        sys.exit(0)
    glutPostRedisplay()           

def draw(*args, **kwargs):
    glClear(GL_COLOR_BUFFER_BIT | GL_DEPTH_BUFFER_BIT) 
    glRotated(0.125,1,1,1)
    R1 = 0.5
    R2 = 0.2
    
    glBegin(GL_QUAD_STRIP)
    glColor3f(1,1,1)
    for j in range(-9,8):
        for i in range(21):
            x1 = (R1 + R2*cos(pi*j/18))*cos(pi*i/10)
            x2 = (R1 + R2*cos(pi*(j+1)/18))*cos(pi*(i+1)/10)
            y1 = (R1 + R2*cos(pi*j/18))*sin(pi*i/10)
            y2 = (R1 + R2*cos(pi*(j+1)/18))*sin(pi*(i+1)/10)
            z1 = R2*sin(pi*j/16)
            z2 = R2*sin((j+1)/16)
            glVertex3d(x1, y1, z1)
            glVertex3d(x2, y2, z2)
    glEnd()

    glutSwapBuffers()          
    glutPostRedisplay()         

glutInitDisplayMode(GLUT_DOUBLE | GLUT_RGB | GLUT_DEPTH)
glutInitWindowSize(800, 600)
glutInitWindowPosition(50, 50)
glutInit(sys.argv)
glutCreateWindow(b"OpenGL")
glutDisplayFunc(draw)
glutKeyboardFunc(keyboardkeys)
init()
glutMainLoop()
```
